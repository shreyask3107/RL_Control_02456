```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
def get_df(path):
    log_data=open(path,'r')
    result={}
    i=0
    for line in log_data:
        line = line.strip()
        line = line.strip('{}')

        columns = line.split(', ') #or w/e you're delimiter/separator is
        data={}
        for c in columns:
            key = c.split(':')[0]
            value=c.split(':')[1]
            data[key.strip('""')]=value
        result[i]=data
        i+=1
    j=json.dumps(result)

    df=pd.read_json(j, orient='index')
    return df
```
